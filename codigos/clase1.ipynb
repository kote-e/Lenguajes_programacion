# Variables y Tipos de Datos

Las variables son espacios en la memoria que almacenan datos  
(en Python no es necesitas declarar el tipo de variable se asigna el tipo automáticamente según lo que guardemos.)  

**Tipos de datos:**

| Tipo            | Ejemplo      | Que es?                                                                      |
| --------------  | ------------ | -----------------------------------------------------------------------------|
| Entero (int)    | 5            | Son basicamente numeros                                                      |  
| Flotante (float)| 3.14         | Permite usar numeros decimales                                               |
| Cadena (str)    | "Hola"       | Nos permite crear cadenas de texto                                           |
| Booleano (bool) | True / False | Es un tipo de dato que trabaja con dos estados, True y False                 |
| Listas          | [1, 2, 3, 4] | Nos permite crear cadenas de lo que sea, es el tipo mas perro bomba en pyhton|
|Dicionario       |              | Nos permite crear un objeto que usa llaves y valores                         |

````python
print("int")
x = 1
y = 2
z = 200
print(x + y + z)

print("---------------------")

print("float")
f1 = 3.14
print(f1)
foo = x + f1
print(type(foo)) # el tipo de dato de foo es float porque al sumar un int con un float el resultado es un float

print("---------------------") 

print("str")
s1 = 'Hola mundo!'
print(s1[5:])
print(s1[6:8])
s2 = s1 + '!!'
print(s2)
print(f'El largo del string es: {len(s2)}')

print("---------------------")

print("bool")
bo= True
print(bool(3))
print(bool([]))
print(bo)

print("---------------------")

print("list")
lista1 = [1, True, [], 'jhsdgfkhjads', {}, ()] # lista heterogenea
print(lista1[3][5])

lista2 = [] # lista vacia
lista2.append('azul') # append agrega un elemento al final de la lista
lista2.append('rojo')
lista2.append('verde')
print(lista2)
lista2.insert(2, 'naranjo') # insert agrega un elemento en la posicion que le indiquemos, desplazando el resto de los elementos a la derecha
print(lista2)
lista2.pop()   # pop elimina el ultimo elemento de la lista
lista2.remove('naranjo') # remove elimina el primer elemento que coincida con el valor que le indiquemos, si no encuentra el valor lanza un error
print(lista2)
colores = lista2
mascotas = ['manjar', 'gallta', 'chubi']
cervezas = ['Becker', 'Royal', 'Cristal']
super_lista = [cervezas, [colores + mascotas]]  # super_lista = [['Becker', 'Royal', 'Cristal'], [['azul', 'rojo', 'verde'], ['manjar', 'gallta', 'chubi']]]

print("---------------------")

print("dict")
dict1 = {
    'colores': colores,
    'mascotas': mascotas,
}

print(dict1)

dict1['cervezas'] = cervezas

for key, value in dict1.items():
    print(f'La lista en la que estoy es: {key}')
    foo = input(f'Agregue algo a la lista {key}')
    dict1[key].append(foo)
    print(f'La lista {key} ahora es: {dict1[key]}')
    print(dict1['cervezas'][2])

print("---------------------")
````

##  Condicionales
Los condicionales permiten ejecutar bloques de código solo si se cumplen ciertas condiciones.  


**Sintaxis básica:**  
```
if condicion1:  
    # código que se ejecuta si la condicion1 se cumple.

elif condicion2: 
    # código si la  condicion 2 se cumple
    
else: #Se usa si ninguna de las condiciones se cumple.
    # código cuando ninguna de las condiciones anteriores se cumple
```

* No siempre se utilizan todos los condicionales, puedes ocupar solo un if si tienes 1 condicion.

En las condicines se pueden utilizar operadores logicos, que permitir construir expreciones que reepresentan condiciones complejas como estas:  
Podemos usar estos operadores para crear condiciones más complejas:

| Operador | Significado        | Ejemplo             |
|----------|--------------------|---------------------|
| and      | Y lógico           | x > 0 and x < 10    |
| or       | O lógico           | x == 0 or y == 0    |
| not      | Negación lógica    | not x == y          |


````python
if True:
    print(1)

print("---------------------")

var = [1, 2, 3, 4] 

if var:
    print("la variable no esta vacia")

print("---------------------")

var1 = ' '
var2 = True
if var1 and var2:
    print(':)')

print("---------------------")


var1 = ''
var2 = True

if var1 and var2 and True:
    print(':)')

print("---------------------")

ar1 = ''
var2 = True

if var1 or var2:
    print(':)')

print("---------------------")

flag = False
if not flag:
    print('uwu')
````

# Ciclos:
permiten repetir un bloque de código varias veces. Hay dos tipos principales:

*Tipos de Ciclos:* 

-while: Repite el bloque de código mientras la condición sea verdadera.  
-for: Itera sobre una secuencia (como una lista, cadena, o rango).  

-el for lo usamos cuando sabemos cuantas veces debo iterar, pero antes, veremos range()

### Range()
Me permite crear un rango de manera rapida


````python
# opcion 1
r1 = range(100)

# opcion 2
r2 = range(0, 101)

# opcion 3, la tercera opcion es el intervalo de espaciado
r3 = range(2, 102, 2)

print(r1)
print(type(r1))

r_nuevo = list(r1)
print(r_nuevo)

r4 = list(range(50, 1050, 50))
print(r4)

print("---------------------")

frutas = ["manzana", "banana", "cereza"]
for fruta in frutas:
    print(fruta)

print("---------------------")

contador = 0
while contador < 3:
    print("Contador:", contador)
    contador += 1
````
