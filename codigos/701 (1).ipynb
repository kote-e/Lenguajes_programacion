```python
def mostrar_matriz(matriz):
    for fila in matriz:
        print(fila)

def cambiar_fila_completa(matriz, indice, caracter = 0):
    matriz[indice] = [caracter for i in range(len(matriz[-1]))]

def cambiar_columna_completa(matriz, indice, caracter = 0):
    for fila in matriz:
        fila[indice] = caracter


matriz = []
N = 9

for i in range(N):
    fila = []
    for j in range(N):
        fila.append(1)

    matriz.append(fila)

mostrar_matriz(matriz)    

```

```python
# matriz[-1] = [0 for i in range(len(matriz[-1]))]
cambiar_fila_completa(matriz, -1)
mostrar_matriz(matriz)
```

```python
# for fila in matriz:
#     fila[0] = 0
cambiar_fila_completa(matriz, 0)
cambiar_columna_completa(matriz, 0, 6)
mostrar_matriz(matriz)
```

```python
pos = 0 
pos2 = -1
for fila in matriz:
    fila[pos] = 0
    fila[pos2] = 0
    pos += 1
    pos2 -= 1

mostrar_matriz(matriz)
```

```python
medio = len(matriz) // 2 
pivot = 0

for e,fila in enumerate(matriz):
    if pivot == 0 :
        pivot +=1 
        continue

    fila[:pivot] = [7 for i in range(pivot)]

    if e >= medio:
        pivot -= 1
    else:
        pivot += 1

mostrar_matriz(matriz)

```
