### ¿Dónde quedamos?

La clase pasada aprendimos que:
- Una clase **hija** hereda todo de la clase **padre** con `class Hija(Padre)`
- Usamos `super().__init__(...)` para no repetir la inicialización del padre
- Podemos **sobreescribir** métodos del padre en la hija
- `isinstance(obj, Clase)` nos dice si un objeto pertenece a una clase (o a alguna de sus hijas)

Hoy vamos a reforzar todo eso con un ejemplo nuevo y luego ustedes van a practicar solos.

### El escenario: un juego de RPG

En un RPG todos los personajes comparten cosas en común:
- Tienen **nombre**, **vida** y **nivel**
- Pueden **atacar** y **recibir daño**
- Pueden **morir** si su vida llega a 0

Pero cada **clase** de personaje tiene sus particularidades:

| | Guerrero | Mago | Arquero |
|---|---|---|---|
| Atributo extra | `furia` | `mana` | `flechas` |
| Ataque especial | Golpe brutal | Bola de fuego | Disparo preciso |

Vamos a modelar esto con herencia.

### Paso 1: La clase base `Personaje`

Aquí ponemos **solo lo que es común a todos**. Nada más.

```python
class Personaje:
    def __init__(self, nombre, vida, nivel):
        self.nombre = nombre
        self.vida   = vida
        self.nivel  = nivel
        self.vivo   = True

    def __str__(self):
        estado = f"❤️  vida: {self.vida}" if self.vivo else "💀 muerto"
        return f"{self.nombre} (nivel {self.nivel}) — {estado}"

    def recibir_danio(self, cantidad):
        self.vida -= cantidad
        if self.vida <= 0:
            self.vida = 0
            self.vivo = False
            print(f"  💀 {self.nombre} ha caído!")

    def atacar(self, objetivo):
        danio = self.nivel * 2
        print(f"  ⚔️  {self.nombre} ataca a {objetivo.nombre} por {danio} de daño")
        objetivo.recibir_danio(danio)

# Probemos
p = Personaje("Genérico", 100, 3)
enemigo = Personaje("Slime", 10, 1)
print(p)
p.atacar(enemigo)
print(enemigo)
```

### Paso 2: La clase `Guerrero`

El guerrero es un `Personaje` pero con `furia` y un ataque especial más poderoso.

```python
class Guerrero(Personaje):
    def __init__(self, nombre, vida, nivel):
        super().__init__(nombre, vida, nivel)  # hereda los atributos del padre
        self.furia = 0                         # atributo propio del guerrero

    def __str__(self):
        base = super().__str__()               # reutiliza el __str__ del padre
        return f"[Guerrero] {base} | furia: {self.furia}"

    def atacar(self, objetivo):
        # el guerrero acumula furia con cada ataque
        super().atacar(objetivo)               # primero hace el ataque normal
        self.furia += 10
        print(f"  🔥 {self.nombre} acumula furia! ({self.furia})")

    def golpe_brutal(self, objetivo):
        # habilidad especial: consume toda la furia como daño extra
        if self.furia == 0:
            print(f"  {self.nombre} no tiene furia suficiente!")
            return
        danio = self.furia
        self.furia = 0
        print(f"  💥 {self.nombre} usa GOLPE BRUTAL por {danio} de daño!")
        objetivo.recibir_danio(danio)

# Probemos
guerrero = Guerrero("Ragnar", 150, 5)
objetivo = Personaje("Goblin", 80, 2)

print(guerrero)
guerrero.atacar(objetivo)
guerrero.atacar(objetivo)
guerrero.golpe_brutal(objetivo)
print(objetivo)
```

### Paso 3: Polimorfismo

> **Polimorfismo** = el mismo método (`atacar`) se comporta distinto según el tipo de objeto.

Lo más poderoso de la herencia es que podemos tratar objetos distintos de forma **uniforme**.

```python
def simular_ronda(lista_personajes, objetivo):
    """Hace que todos en la lista ataquen al objetivo en una ronda."""
    print(f"\n=== RONDA — objetivo: {objetivo.nombre} (vida: {objetivo.vida}) ===")
    for p in lista_personajes:
        if p.vivo:
            p.atacar(objetivo)   # cada uno ataca a su manera

# Mezclamos tipos distintos en una misma lista
guerrero2 = Guerrero("Astrid", 120, 4)
generico   = Personaje("Aldeano", 60, 1)

grupo = [guerrero, guerrero2, generico]
jefe  = Personaje("Dragon", 200, 10)

simular_ronda(grupo, jefe)
print("\n", jefe)
```
