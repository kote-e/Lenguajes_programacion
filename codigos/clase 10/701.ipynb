## Repaso: el setter

El punto que más confunde es entender **cuándo se llama el setter y cuándo no**.

Regla simple:

```
objeto.attr = valor   →  llama al setter   (usa self.attr)
objeto._attr = valor  →  escribe directo   (salta el setter)
```

Y dentro del `__init__`, la misma regla aplica — depende de si escribes con o sin `_`:

````python
class Temperatura:
    def __init__(self, valor):
        self._grados = 0        # escribe directo, sin validar
        self.grados  = valor    # pasa por el setter — aquí SÍ se valida

    @property
    def grados(self):
        return self._grados

    @grados.setter
    def grados(self, valor):
        print(f"  setter llamado con: {valor}")
        if valor < -273.15:
            print("  ❌ bajo cero absoluto, ignorado")
            return
        self._grados = valor    # solo aquí se guarda

print("--- creando objeto ---")
t = Temperatura(25)
print(f"grados: {t.grados}\n")

print("--- asignando con setter ---")
t.grados = -300     # setter bloqueado
t.grados = 100      # setter acepta
print(f"grados: {t.grados}\n")

print("--- saltando el setter ---")
t._grados = -300    # escribe directo sin validar
print(f"grados: {t.grados}")
````

El `print` dentro del setter permite ver exactamente en qué momento se activa. En producción no se deja, pero para entender el flujo es útil.
