## ¿Qué es recursión?

Una función es **recursiva** cuando se llama a sí misma para resolver una versión más pequeña del mismo problema.

La idea central: *si puedo resolver el problema chico, puedo resolver el problema grande.*

Toda función recursiva tiene exactamente dos partes obligatorias:

```
1. Caso base   → el problema más pequeño posible, se resuelve directo (sin llamarse)
2. Caso recursivo → reduce el problema y se llama a sí mismo
```

Sin caso base, la función se llama infinitamente hasta que Python corta con un error.

---

## Primer ejemplo: cuenta regresiva

Problema: imprimir los números desde `n` hasta `0`.

Con un `while` es trivial. ¿Cómo lo pienso de forma recursiva?

> *"Imprimir desde n hasta 0" es lo mismo que imprimir n, y luego imprimir desde n-1 hasta 0.*

````python
def cuenta_regresiva(n):
    if n < 0:          # caso base: no hay nada que imprimir
        return
    print(n)
    cuenta_regresiva(n - 1)   # caso recursivo: problema más pequeño

cuenta_regresiva(5)
````

### Ver la recursión en acción

Podemos agregar una **sangría** proporcional a la profundidad para visualizar cómo la función se llama a sí misma y cómo regresa.

````python
def cuenta_regresiva(n, profundidad=0):
    sangria = "  " * profundidad
    print(f"{sangria}→ llamada con n={n}")

    if n < 0:
        print(f"{sangria}  caso base alcanzado")
        return

    print(f"{sangria}  {n}")
    cuenta_regresiva(n - 1, profundidad + 1)
    print(f"{sangria}← regresando de n={n}")

cuenta_regresiva(3)
````

La función baja hasta el caso base y luego **regresa** por cada llamada pendiente. Eso es la **pila de llamadas** (call stack).

## Segundo ejemplo: factorial

```
5! = 5 × 4 × 3 × 2 × 1
   = 5 × 4!
```

El factorial de `n` es `n` multiplicado por el factorial de `n-1`. El caso base es `0! = 1`.

````python
def factorial(n):
    if n == 0:                      # caso base
        return 1
    return n * factorial(n - 1)     # caso recursivo

for i in range(8):
    print(f"{i}! = {factorial(i)}")
````

````python
# Para verlo con la pila
def factorial(n, prof=0):
    s = "  " * prof
    print(f"{s}factorial({n})")
    if n == 0:
        print(f"{s}→ retorna 1")
        return 1
    resultado = n * factorial(n - 1, prof + 1)
    print(f"{s}→ retorna {n} × ... = {resultado}")
    return resultado

factorial(4)
````

## Tercer ejemplo: potencia

¿Cómo calcular `base ** exponente` de forma recursiva?

```
2^5 = 2 × 2^4
2^4 = 2 × 2^3
...
2^0 = 1   ← caso base
```

````python
def potencia(base, exp):
    if exp == 0:                          # caso base
        return 1
    return base * potencia(base, exp - 1) # caso recursivo

print(potencia(2, 10))
print(potencia(3, 4))
````

## El ejemplo clásico: Torres de Hanói

Hay tres postes: **A**, **B** y **C**. En el poste A hay `n` discos apilados de mayor a menor. El objetivo es mover todos los discos al poste C, con dos reglas:

1. Solo se mueve un disco a la vez.
2. Nunca se puede poner un disco más grande sobre uno más pequeño.

### ¿Por qué es recursivo?

Para mover `n` discos de A a C usando B como auxiliar:

1. Mueve los `n-1` discos de encima de A a B (usando C como auxiliar)
2. Mueve el disco grande de A a C
3. Mueve los `n-1` discos de B a C (usando A como auxiliar)

El caso base: mover 1 disco es trivial — simplemente se mueve.

````python
def hanoi(n, origen, destino, auxiliar):
    if n == 1:                                          # caso base
        print(f"  Disco 1: {origen} → {destino}")
        return
    hanoi(n - 1, origen, auxiliar, destino)             # paso 1
    print(f"  Disco {n}: {origen} → {destino}")         # paso 2
    hanoi(n - 1, auxiliar, destino, origen)             # paso 3

print("Torres de Hanói con 3 discos:")
hanoi(3, "A", "C", "B")
````

````python
# ¿Cuántos movimientos necesita?
movimientos = 0

def hanoi_contar(n, origen, destino, auxiliar):
    global movimientos
    if n == 1:
        movimientos += 1
        return
    hanoi_contar(n - 1, origen, auxiliar, destino)
    movimientos += 1
    hanoi_contar(n - 1, auxiliar, destino, origen)

for discos in range(1, 11):
    movimientos = 0
    hanoi_contar(discos, "A", "C", "B")
    print(f"  {discos} discos → {movimientos:5} movimientos  (2^{discos} - 1 = {2**discos - 1})")
````

El número de movimientos es siempre `2^n - 1`. Con 64 discos (el problema original de la leyenda) serían **18.446.744.073.709.551.615** movimientos.

---

## Resumen

| | Qué hace |
|---|---|
| Caso base | Corta la recursión — sin él es infinita |
| Caso recursivo | Llama a la función con un problema más pequeño |
| Pila de llamadas | Cada llamada queda suspendida hasta que la siguiente retorna |

La recursión no es mejor ni peor que un loop — es una forma distinta de pensar el problema. Algunos problemas se expresan de forma mucho más natural de forma recursiva (Hanói, árboles, fractales).
